Previso de demanda  uma projeo realizada com o intuito de prever o potencial de vendas de um ou mais produtos em um cenrio futuro.

Ok! Vimos que o melhor modelo  o ***dt_cds_dt	Decision Tree w/ Cond. Deseasonalize & Detrending***. Ento iremos aplic-lo.

```python
# Atribuindo o modelo  uma varivel de mesmo nome
dt_cds_dt = create_model('dt_cds_dt')
```
